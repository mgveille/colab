```python
import pandas as pd
url ='https://raw.githubusercontent.com/mgveille/colab/main/cac_allshare_mars2023.csv'
df_cactickers = pd.read_csv(url)
df_cactickers.head()
```

```python
ls_cactickers = df_cactickers['ticker'].tolist()
print(ls_cactickers)
```

```python
import yfinance as yf
df_cacstocks = yf.download(ls_cactickers , period= '1mo')['Close']
# yf.download(line,period = '10y', interval='1d')['Close']
# df = yf.Ticker("IBM").history(start="2017-01-01", end="2017-04-30", frequency='1dy')['Close']
```
